```python
import pandas as pd
import numpy as np
import datetime
```

```python
baseData = pd.read_csv('201710-citibike-tripdata.csv', delimiter=',')
```

```python
baseData.head()
```

```python
len(baseData)
```

```python
baseData.dtypes
```

```python
baseData['starttime'] = pd.to_datetime(baseData['starttime'])
```

```python
baseData.dtypes
```

```python
selectedTrips = baseData[(baseData['starttime'] >= datetime.date(2017, 10, 11)) & (baseData['starttime'] < datetime.date(2017, 10, 12))]
```

```python
selectedTrips
```

```python
selectedTrips = selectedTrips[selectedTrips['birth year'] > 0]
```

```python
selectedTrips
```

```python
selectedTrips.sample(n=5000).to_csv('20171011_SelectedCitibikeTrips_Rnd5000.csv')
```
